```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 1, 50)
y = 2*x + 1
```

```python
plt.figure()
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
```

```python
plt.figure()
plt.plot(x, y1)
plt.show()
```

```python
plt.figure(num=3, figsize=(8,5),) # 编号3，大小为8,5
# zorder调整显示顺序
l1, = plt.plot(x, y1, label='linear line', linewidth=10, zorder=1)
l2, = plt.plot(x, y2, 
    color='red', linewidth=1.0, linestyle='--', label='square line')
# plt.legend(loc='upper right') # 显示标签
plt.legend(handles=[l1, l2], labels=['up', 'down'], loc='best')

x0 = 1 # 要标注的x0和y0
y0 = 2*x0 + 1
plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5) # 绘制虚线
plt.scatter([x0,], [y0,], s=50, color='b') # 绘制着重强调的点

# xycoords 基于数据选位置
# xytext 标注的位置
# textcorrds 偏差值
plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data',
    textcoords='offset points', xytext=(+30, -30), fontsize=16,
    arrowprops=dict(arrowstyle='->', 
                    connectionstyle='arc3, rad=.2'))

# 注释
plt.text(-1, 4, 
         r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',
         fontdict={'size': 12, 'color': 'r'}) # 字号和颜色

plt.xlim((-1, 2)) # X轴范围
plt.ylim((-2, 6)) # Y轴范围
plt.xlabel('I am x') # X轴标签
plt.ylabel('I am y') # Y轴标签
new_ticks = np.linspace(-1, 2, 10) 
plt.xticks(new_ticks) # 标注点
plt.yticks([-2, -1.8, -1, 1.22, 3],[r'$really\ bad$', # 标注点加名字
    r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])

ax = plt.gca() # 获取坐标轴信息
ax.xaxis.set_ticks_position('bottom') # X轴标注点显示在底部
ax.spines['right'].set_color('none') # 设置右边框颜色为无
ax.spines['top'].set_color('none') # 设置上边框颜色为无
ax.spines['bottom'].set_position(('data', 0)) # 设置X轴位置
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

# 刻度数字设置透明度
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white', edgecolor='None',
        alpha=0.7, zorder=2))

plt.show()
```

```python
n = 1024
# 标准正态分布，平均数为0，方差为1
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)
```

```python
plt.scatter(X, Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5, 1.5)
plt.xticks(())
plt.ylim(-1,5, 1,5)
plt.yticks(())

plt.show()
```

```python
n = 12
X = np.arange(n)
# 随机分布
Y1 = (1 - X/float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X/float(n)) * np.random.uniform(0.5, 1.0, n)

# 加颜色
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

plt.xlim(-1, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
plt.yticks(())

# 保留两位小数，横向居中对齐，纵向底部对齐
for x,y in zip(X, Y1):
    plt.text(x, y+0.05, '%.2f' % y, ha='center', va='bottom')

for x,y in zip(X, Y2):
    plt.text(x, -y-0.05, '%.2f' % y, ha='center', va='top')
    
plt.show()
```

```python
def f(x, y):
    # the height function
    return (1-x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y) # 编织成栅格
```

```python
# 颜色填充 f(X, Y)映射到暖色组
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)
# 划线 8代表等高线密集程度
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
# 标签
plt.clabel(C, inline=True, fontsize=10)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
# 在窗口上添加3D坐标轴
ax = Axes3D(fig)
# X,Y,Z
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y) # 编织网格
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)

# stride 跨度
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, 
                cmap=plt.get_cmap('rainbow'))

# 投影
ax.contourf(X, Y, Z, zdir='z', offset=-2, 
            cmap=plt.get_cmap('rainbow'))
ax.set_zlim(-2, 2)

plt.show()
```

```python
fig = plt.figure()
x = [1,2,3,4,5,6,7]
y = [1,3,4,2,5,8,6]
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
# 绘制大图
ax1 = fig.add_axes([left, bottom, width, height])
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

left, bottom, width, height = 0.2, 0.6, 0.25, 0.25
# 绘制小图
ax2 = fig.add_axes([left, bottom, width, height])
ax2.plot(y, x, 'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title inside 1')

# 另一种方法
plt.axes([0.6, 0.2, 0.25, 0.25])
plt.plot(y[::-1], x, 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title inside 2')

plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y1 = 0.05 * x**2
y2 = -1 * y1
fig, ax1 = plt.subplots()

ax2 = ax1.twinx() # 生成ax1的镜面
ax1.plot(x, y1, 'g-')
ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data', color='g')
ax2.plot(x, y2, 'b-')
ax2.set_ylabel('Y2 data', color='b')
plt.show()
```

```python
from matplotlib import animation
fig, ax = plt.subplots()
x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i/10.0))
    return line,
def init():
    line.set_ydata(np.sin(x))
    return line,

ani = animation.FuncAnimation(fig=fig, func=animate, frames=100,
    init_func=init, interval=20, blit=False)

plt.show()
```
