```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DNA_SIZE = 10 # DNA长度
POP_SIZE = 100 # 种群数量
CROSS_RATE = 0.8 # DNA重组概率
MUTATION_RATE = 0.003 # 变异程度
N_GENERATIONS = 200 # 繁殖代数
X_BOUND = [0, 5] # X的取值范围
```

```python
def F(x): return np.sin(10*x)*x + np.cos(2*x)*x # 目标函数

# 返回环境适应程度
def get_fitness(pred): return pred + 1e-3 - np.min(pred)

# 将二进制DNA转为十进制，正则化至（0,5）
def translateDNA(pop): 
    return pop.dot(
        2 ** np.arange(
            DNA_SIZE)[::-1])/float(2**DNA_SIZE-1) * X_BOUND[1]

def select(pop, fitness): # 自然选择
    # p 就是被选中的概率
    idx = np.random.choice(np.arange(POP_SIZE), size=POP_SIZE,
        replace=True, p=fitness/fitness.sum())
    return pop[idx]

def crossover(parent, pop): # 基因重组
    if np.random.rand() < CROSS_RATE:
        # 从种群中选取另一个个体
        i_ = np.random.randint(0, POP_SIZE, size=1)
        # 选择交换的基因点
        cross_points = np.random.randint(
            0, 2, size=DNA_SIZE).astype(np.bool)
        parent[cross_points] = pop[i_, cross_points]
    return parent
        
# 变异
def mutate(child):
    for point in range(DNA_SIZE):
        if np.random.rand() < MUTATION_RATE:
            child[point] = 1 if child[point] == 0 else 0
        return child
```

```python
pop = np.random.randint(2, size=(POP_SIZE, DNA_SIZE)) # 初始化种群
```

```python
plt.ion()
x = np.linspace(*X_BOUND, 200)
plt.plot(x, F(x))
```

```python
for _ in range(N_GENERATIONS):
    F_values = F(translateDNA(pop)) # 计算DNA的值
    
    if 'sca' in globals():
        sca.remove()
    sca = plt.scatter(translateDNA(pop), F_values, s=200, lw=0,
        c='red', alpha=0.5)
    plt.pause(0.05)
    
    # 进化
    fitness = get_fitness(F_values)
    print('Most fitted DNA: ', pop[np.argmax(fitness), :])
    pop = select(pop, fitness)
    pop_copy = pop.copy()
    for parent in pop:
        child = crossover(parent, pop_copy)
        child = mutate(child)
        parent[:] = child # 父代被子代替代
        
plt.ioff()
plt.show()
```
